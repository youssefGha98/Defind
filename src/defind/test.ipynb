```python
import pandas as pd
from web3 import Web3

# Pool ABI - just the gauge() function
POOL_ABI = [
    {
        "constant": True,
        "inputs": [],
        "name": "gauge", 
        "outputs": [{"name": "", "type": "address"}],
        "payable": False,
        "stateMutability": "view",
        "type": "function"
    }
]

def get_gauge_for_pool(w3: Web3, pool_addr: str) -> str:
    """Get gauge address for a given pool address."""
    try:
        pool = w3.eth.contract(address=Web3.to_checksum_address(pool_addr), abi=POOL_ABI)
        return pool.functions.gauge().call()
    except Exception as e:
        print(f"Error getting gauge for pool {pool_addr}: {e}")
        return None

w3 = Web3(Web3.HTTPProvider("https://optimism-mainnet.chainnodes.org/6457c046-a111-448d-8997-3e164fe1149b"))

input_path = "data2/0x548118c7e0b865c2cfa94d15ec86b666468ac758__topics-0xab0d57f0__universal/shards/shard_00001.parquet"

df = pd.read_parquet(input_path)

pool_column = "pool"  


unique_pools = df[pool_column].unique()

# Create mapping
pool_to_gauge = {}
for pool_addr in unique_pools:
    pool_to_gauge[pool_addr] = get_gauge_for_pool(w3, pool_addr)

# Add gauge column
df['gauge_address'] = df[pool_column].map(pool_to_gauge)

# # Save result
# output_path = "data2/0x548118c7e0b865c2cfa94d15ec86b666468ac758__topics-0xab0d57f0__universal/shards/shard_00001_with_gauges.parquet"
# df.to_parquet(output_path, index=False)
# print(f"ðŸ’¾ Saved enhanced data to: {output_path}")

# Show results
successful = df['gauge_address'].notna().sum()

```
