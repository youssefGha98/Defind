```python
import duckdb

# target wallet (lowercase for consistency)

# open a DuckDB connection (in-memory)
con = duckdb.connect()
con.execute("PRAGMA memory_limit='2GB'")  # optional, bound memory

# query across all shards
q = """
SELECT block_number,	block_timestamp,	tx_hash,	log_index,	contract,	event,	amount0,	amount1,	liquidity,	token_id
FROM read_parquet(
  'data2/0x416b433906b1b72fa758e166e239c43d68dc6f29__topics-0x26f6a048x0x3067048b__universal/shards/shard_*.parquet',
  union_by_name=true
)
WHERE token_id = ?
ORDER BY block_number, tx_hash, log_index
"""

df = con.execute(q,["3704259"]).df()
print(df.head())
```
